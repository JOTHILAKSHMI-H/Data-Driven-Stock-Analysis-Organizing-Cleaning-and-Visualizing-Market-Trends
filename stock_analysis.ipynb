```python
import pandas as pd
import yaml 
import os 

#function to read yaml file
def read_yaml_file(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
        return pd.json_normalize(data)
    
    
stock_combined_df =pd.DataFrame()   
    
#generate the month folder using date_range 
 
months = pd.date_range(start = '2023-11-01', end = '2024-11-22', freq='MS').strftime('%Y-%m').to_list() #2023-11

for month in months:
    folder_path = os.path.join ('E:/stock analysis', month) #E:/stock analysis\\2024-11'
    try:
        
    
        # Generate a list of possible date files
        month_start = pd.to_datetime(f"{month}-01" )#month 2023-11-01, string formated date into datetime
        month_end = month_start + pd.offsets.MonthEnd(1)# current month last date
        date_files = pd.date_range(start = month_start, end = month_end).strftime('%Y-%m-%d').to_list()

        for date_file in date_files:
            file_name = f"{date_file}_05-30-00.yaml"
            file_path = os.path.join(folder_path, file_name)
            
            try:
                df = read_yaml_file(file_path)
                stock_combined_df = pd.concat([stock_combined_df, df], ignore_index=True) #list of dataframe is concate
                
            except FileNotFoundError:
                print(f"file{file_path}not found.")
            except Exception as e:
                print(f"An error occurred while reading {file_path}:{e}")   
                
    except FileNotFoundError:
        print(f"file{file_path}not found.")
    except Exception as e:
        print(f"An error occured while reading {folder_path}:{e}")   
        
print(stock_combined_df) 
              
```

```python
def correlation_matrix_df(correlation_matrix):
    correlations = []
    
    for row in correlation_matrix.index:
        for col in correlation_matrix.columns:
            if row !=col:
                correlations.append({'stock_1':row, 'stock_2':col, 'correlation':correlation_matrix.loc[row, col] })
                correlation_df = pd.DataFrame(correlations)
    return correlation_df
    
```

```python
correlation_df = correlation_matrix_df(correlation_matrix)
```

```python
correlation_df
```

```python
correlation_df.info()
```

```python
def top_correlation(correlation_df, top_corr = 10):
    correlation_df= correlation_df[correlation_df['correlation'] >0 ]
    sorted_correlation_df = correlation_df.sort_values(by = 'correlation', ascending= False)
    sorted_correlation_df = sorted_correlation_df.drop_duplicates(subset = ['correlation'])
    return sorted_correlation_df.head(top_corr)
     
```

```python
top_10_correlation = top_correlation(correlation_df, top_corr = 10).reset_index(drop = True)
```

```python
#connect sql
import pymysql
import pandas as pd

db_connection = pymysql.connect(
    host = "localhost",
    database = 'stock_analysis',
    user = 'mat2',
    password ='0701'
)
```

```python
curr = db_connection.cursor()
```

```python
combined_stock_csv.info()
```

```python
#create the table
create_table_query = """
CREATE TABLE IF NOT EXISTS stock_analysis(
    id INT AUTO_INCREMENT PRIMARY KEY,
    Ticker VARCHAR(250),
    close FLOAT,
    date DATE,
    high FLOAT,
    low FLOAT,
    month VARCHAR(7),
    open FLOAT,
    volume BIGINT,
    year SMALLINT,
    daily_return FLOAT,
    company VARCHAR(250),
    sector VARCHAR(250),
    yearly_return FLOAT,
    monthly_return FLOAT
)"""
curr.execute(create_table_query)
db_connection.commit()
```

```python
#insert the table into the query
insert_query = """
INSERT INTO stock_analysis(Ticker, close, date, high, low, month, open, volume, year, daily_return, company, sector, yearly_return, monthly_return) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""
```

```python
df = combined_stock_csv
```

```python
df
```

```python
# Iterate over the DataFrame and insert each row
for index, row in df.iterrows():
    curr.execute(insert_query, (row['Ticker'],
                                row['close'],
                                row['date'],
                                row['high'],
                                row['low'], 
                                row['month'],
                                row['open'],
                                row['volume'],
                                row['year'],
                                row['daily_return'],
                                row['company'],
                                row['sector'],
                                row['yearly_return'],
                                row['monthly_return']))
    

db_connection.commit() 
```

```python
# Run a SELECT query to fetch all rows
select_query = "SELECT * FROM stock_analysis"
curr.execute(select_query)

# Fetch all rows from the result of the query
rows = curr.fetchall()

db_connection.commit()
```

```python
#create a table for #top 10 volatility
create_table_query = """CREATE TABLE IF NOT EXISTS top_10_volatility
(
    Ticker Varchar(250),
    volatility Float
)
"""
curr.execute(create_table_query)

#
top_10_volatility_values =  top_10_volatility.values.tolist()

#insert the table
insert_query = """ insert into top_10_volatility(Ticker, volatility) values(%s, %s)"""
curr.executemany(insert_query, top_10_volatility_values)
db_connection.commit()
```
